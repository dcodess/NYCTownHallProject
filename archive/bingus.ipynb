```python


# Determine the number of rows and columns for your subplots
n_rows = int(len(train_data.columns) / 5) + 1
n_cols = 5

# Create a figure with a specific size
fig, axes = plt.subplots(n_rows, n_cols, figsize=(75, 50))
axes = axes.ravel()  # Flatten the 2D array of axes for easier iteration

# Iterate through the columns of the DataFrame and plot histograms
for i, column in enumerate(train_data.columns):
    if i >= len(train_data.columns):
        break  # Handle extra spaces in the subplot grid
    ax = axes[i]
    train_data[column].plot(kind='hist', ax=ax)
    ax.set_title(f'Histogram of {column}')
    ax.set_xlabel('Range(Percent of Population) 0.0 to 1.0')
    ax.set_ylabel('Frequency within Boro')
    
    #Gridlines???
    # ax.grid(True)

# Remove any empty subplots
for i in range(len(train_data.columns), n_rows * n_cols):
    fig.delaxes(axes[i])

# Adjust the layout
# plt.tight_layout()

# Display the entire figure
plt.show()
```

```python
#Grid search with Cross validation
# Spliting the data into n folds and training the data then using one fold to test the data

from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()


#using a paramater grid to find the besst possible score
param_grid = {
    "n_estimators": [100, 200, 300],  # Wider range
    "max_features": [2, 4, 6, 8],
    "max_depth": [None, 10, 20, 30],  # Experiment with different depths
    "min_samples_split": [2, 5, 10],  # Vary the minimum samples required to split
    "min_samples_leaf": [1, 2, 4],  # Vary the minimum samples required in a leaf
}

grid_search = GridSearchCV(forest, param_grid, cv=5, scoring="neg_mean_squared_error", return_train_score=True)
grid_search.fit(x_train_s, y_train)


grid_search = GridSearchCV(forest, param_grid, cv=5, scoring = "neg_mean_squared_error",
                           return_train_score= True)

grid_search.fit(x_train_s, y_train)
```
